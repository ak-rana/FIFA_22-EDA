```python
plt.xlabel('Overall score')
plt.ylabel('Value in Euros')
plt.title('Overall Player Score by Value in Euros ')

 
plt.bar(df['overall'],df['value_eur'])
```

It is observed from the data visual representation:
1. There's a direct relation between overall player score and the player worth/value in Euros
2. As the player's overall score increases there value in Euros (standard measure) also increases. 
3. The data is negatively skewed which means most values lies on the right side of the distribution graph.


```python

plt.xlabel('Potential')
plt.ylabel('Wage_Euros')
plt.title('Potential wise wage')

 
plt.scatter(df['potential'],df['wage_eur'])
```

It is observed from the data visual representation:
1. There's a direct relation between the two variables Potential and Wages in Euros
2. Moreover, it is observed that from potential score of 50-65, the player's wages have acute differences. Whereas, it has a significant increase as the potential score increases beyond 65.

```python
sns.relplot(data=df, x="international_reputation", y="age", kind="line")
```

It is observed from the data visual representation:
1. There's a direct relation between the two variables Age and International Reputaion
2. As the age increases, we're aware that the public awareness of a player eventually increases and likely is the international reputation of that player. Evidently, it is indeed depicted in the line graph.

```python
plt.title('Age wise Potential Comparison ')

sns.lineplot(x='age', y='potential', data=df)
plt.show()
```

It is observed from the data visual representation:
1. Potential for a player is at its highest at the young age, as the player ages and becomes more experienced, the potential eventually keeps on declining as skills become more polished & refined. 
2. It is also observed, that there's a steep drop in potential at age 42 approx., reasons being unknown with the in-hand information. 
3. However, potential dropes significantly from age 44, assumed reasons for the same could be aging & retirement.

```python
plt.xlabel('age')
plt.ylabel('Value in Euros')
plt.title('Age wise Value in Euros ')

 
plt.bar(df['age'],df['value_eur'])
```

It is observed from the data visual representation:
1. Players at the age of 21 has been seen to have the maximum value/worth in Euros.
2.  Approximately 90% of data ranges between the age of 20 - 35, hence players worth is maximum during these ages assumed factors could be internationa reputation, experience, etc.
3. The graphical representation doesn't happen to have a evident skewness, and mostly values are concentrated at the centre.
